```python
df = pd.DataFrame({"No": No, "y": y, "x1":x1, "x2":x2,"x3":x3})
df

```

```python
df["x1_squared"] = df ["x1"]**2
df["x2_squared"] = df ["x2"]**2
df["x3_squared"] = df ["x3"]**2
df
```
